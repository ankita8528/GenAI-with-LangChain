### What is LangChain?
LangChain is an open source framework that helps in building LLM based applications. It provides modular components and end to end tools that help developers build comples AI applications, such as *Chatbots*, *Question Answering Systems*, *RAG*, *Autonomous Agents* and more.

1. Supports all the major LLMs.
2. Simplifies develping available LLM based applications.
3. Integrations available for all major tools.
4. Open source/Free/Actively developed.
5. Supports all major Gen AI use cases.

More like LangChain - LlamaIndex, Haystack, etc.

## Why do we need LangChain?
Imagine you want to build a RAG-based system where a user can upload any PDF or document and then ask queries related only to that content. You’d first process the document(s) and convert them into embeddings, which are then stored in a Vector Database instead of saving the original files directly.

Now there are some challanges, when a user asks a question, you need more than just the stored embeddings—you need a framework that can:

1. Understand the user’s query,
2. Semantic search for the relevant portions of the embedded documents,
3. And return a precise response that stays within the context of the user’s question.

This is exactly where LangChain comes in—it acts as the glue connecting your documents, vector database, and the LLM to deliver accurate, context-specific answers. ** LangChain can handle all the inferences that are the challanges**.
